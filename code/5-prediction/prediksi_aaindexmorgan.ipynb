```python
import pandas as pd
from joblib import load
```

```python
testing_df = pd.read_csv('../../data/5-prediction/input/data_prediksi.csv') 
testing_df
```

```python
inter_df = testing_df[['UniProt ID','Drug']]
inter_df
```

```python
testing_dff = testing_df.drop(columns=['UniProt ID','Drug'])
testing_dff
```

```python
testing_dff.columns = range(len(testing_dff.columns))
testing_dff
```

### Prediksi menggunakan Model Random Forest

```python
model = load('../../data/4-modelling/result/bindingdb/AAIndex-Morgan/best_model_random_forest.joblib')
```

```python
print(type(model))
```

```python
hasil_prediksi = model.predict(testing_dff)
hasil_prediksi
```

```python
proba = model.predict_proba(testing_dff)
proba 
```

```python
# Menggunakan .predict_proba() untuk mendapatkan probabilitas prediksi
y_proba = model.predict_proba(testing_dff)[:, 1]  # Probabilitas untuk kelas 1

# Menetapkan threshold 0.9 untuk prediksi kelas 1
threshold = 0.95
y_pred_custom = (y_proba >= threshold).astype(int)

print("Probabilitas prediksi untuk kelas 1:", y_proba)
print("Prediksi dengan threshold 0.8:", y_pred_custom)
```

```python
hasil = pd.DataFrame(y_proba)
hasil
```

```python
result = pd.concat([inter_df, hasil], axis=1)
result
```

```python
df_sorted = result.sort_values(by=0, ascending=False)
# df_sorted = df_sorted.head(10)
df_sorted
```

```python
filter1_df = df_sorted[df_sorted[0] > 0.95]
filter1_df
```

```python

# Membuat distribusi normal dari kolom
# Kita melakukan standarisasi nilai menjadi distribusi normal dengan mean 0 dan std 1
mean_value = df_sorted[0].mean()
std_value = df_sorted[0].std()
df_sorted['Z_Score'] = (df_sorted[0] - mean_value) / std_value

# # Alternatif lain menggunakan scipy.stats.zscore
# df['Z_Score_Alt'] = zscore(df['Values'])

# Output hasil
df_sorted
```

```python
mean_value
```

```python
std_value
```

```python
filtered_df1 = df_sorted[df_sorted["Z_Score"] > 1.65]
filtered_df1
```

```python
filtered_df = df_sorted[df_sorted[0] > 0.9]
filtered_df
```

```python
df_sorted1 = result.sort_values(by=0, ascending=False)
df_sorted1 = df_sorted1.head(10)
df_sorted1
```

```python
from scipy.stats import norm
# p-value untuk one-tailed (right)
df_sorted['p_value'] = 1 - norm.cdf(df_sorted['Z_Score'])

# # Alternatif lain menggunakan scipy.stats.zscore
# df['Z_Score_Alt'] = zscore(df['Values'])

# Output hasil
df_sorted
```

```python
p_value = df_sorted[df_sorted["p_value"] < 0.05 ]
p_value
```

```python
p_value.shape
```

```python
p_value.to_csv('../../data/5-prediction/output/rf_predict_result.csv', index=False)
```

### Jumlah senyawa dan jumlah protein
