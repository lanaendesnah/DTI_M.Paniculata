```python
import pandas as pd
```

```python
neg = pd.read_csv("../../data/3-nrlmf/2-output_nrlmf/neg_inter_chembl.csv")
neg = neg.drop(columns=['score'])
neg['Class'] = 0
neg
```

```python
pos = pd.read_csv("../../data/1-preparation/interaction/interaction_chembl.csv")
pos = pos.drop(columns=['SMILES', 'FASTA Sequence'])
pos['Class'] = 1
pos
```

```python
data = pd.concat([pos, neg])
data = data.reset_index(drop=True)
data
```

```python
aaindex = pd.read_csv('../../data/2-feature/prot/aaindex_nrlmf_chembl.csv')
aaindex
```

```python
morgan = pd.read_csv('../../data/2-feature/mol/morgan_chembl.csv')
morgan
```

```python
df = pd.merge(data, aaindex, on='UniProt ID', how='left')
df
```

```python
df2 = pd.merge(df, morgan, on='Drug', how='left')
df2
```

```python
df2 = df2.drop(columns=['UniProt ID', 'Drug'])
first_col = df2.pop('Class')
df2['Class'] = first_col
df2
```

```python
df2.to_csv('../../data/4-modelling/1-datainput/aaindex-morgan_chembl.csv', index = False)
```
